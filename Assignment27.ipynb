```python
# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

# ans
""" 
In Bokeh, glyphs are visual markers or shapes that represent data points on a plot. 
They can be used to create various types of visualizations, such as scatter plots, line plots, bar plots, 
and more. Bokeh provides a wide range of glyph methods that allow you to add different types of glyphs
to your plots.

To add glyphs to a Bokeh plot, you typically follow these steps:

Create a figure: Use the bokeh.plotting.figure function to create a new figure object. Specify the plot
dimensions, axis labels, and other options as needed.

Add glyphs to the figure: Use the glyph methods provided by Bokeh to add visual elements to the figure.
Each glyph method corresponds to a different type of glyph. Some commonly used glyph methods include:

circle(): Adds circle glyphs to the plot.
line(): Adds line glyphs to the plot.
rect(): Adds rectangle glyphs to the plot.
triangle(): Adds triangle glyphs to the plot.
square(): Adds square glyphs to the plot.
patch(): Adds filled polygons as glyphs to the plot.
Specify data sources: Associate your data with the glyphs by providing the data source argument to the glyph
methods. Bokeh accepts data sources in the form of Pandas DataFrames, NumPy arrays, or Python dictionaries.

Customize glyph appearance: You can customize the appearance of the glyphs by passing additional parameters
to the glyph methods. These parameters control properties like size, color, transparency, line width, and 
more.

Display the plot: Use the bokeh.io.show function to display the plot in a web browser or Jupyter Notebook. """
```

```python
# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

# ans
""" Bokeh provides several options to customize the appearance of a plot, including the axes, title, and
legend. Here are some ways you can customize these elements:

1.Plot Title:
Set the title of the plot using the title property of the figure object.
For example: p.title.text = "My Plot Title"
Customize the appearance of the title using additional properties, such as text_font, text_font_size, 
text_color, text_alpha, etc.

2.Axes:
Customize the x-axis and y-axis labels using the xaxis.label and yaxis.label properties of the figure object. 
For example: p.xaxis.axis_label = "X-axis Label"
Set the appearance of the axis labels using properties like axis_label_text_font, axis_label_text_font_size,
axis_label_text_color, etc.
Adjust the ticks and tick labels on the axes using properties such as axis_label_text_font_size,
major_label_text_font_size, axis_line_color, major_tick_line_color, minor_tick_line_color, etc.
Modify other axis properties like axis_line_width, axis_line_alpha, axis_line_dash, major_tick_line_width,
major_tick_line_alpha, major_tick_line_dash, etc.

3.Legend:
Add a legend to the plot by setting the legend.label property of the glyphs to a meaningful value.
For example: p.circle(x, y, legend_label="Data Points")
Customize the appearance of the legend using properties like legend.title, legend.title_text_font, 
legend.title_text_font_size, legend.title_text_color, legend.title_text_alpha, etc.
Modify other legend properties like legend.label_text_font, legend.label_text_font_size, 
legend.label_text_color, legend.label_text_alpha, etc. """
```

```python
""" Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time? """

# ans
""" The Bokeh server is a Python library that allows you to create interactive plots and applications that
can be updated in real time. It provides a server-side framework for deploying Bokeh applications,
enabling you to create dynamic and interactive visualizations that respond to user input or data updates.

By using the Bokeh server, you can create plots with interactive features such as sliders, buttons,
dropdown menus, and more. These interactive components can trigger updates to the plot based on user 
actions, allowing for real-time exploration and analysis of data.

To use the Bokeh server and create interactive plots, you need to follow these general steps:

Define your plot and interactive components: Create your Bokeh plot, including any interactive components
you want to include, such as sliders, buttons, etc. Set up the necessary callbacks to handle user input or
data updates.

Create a Bokeh application: Wrap your plot and interactive components in a Bokeh application using the
bokeh.application module. This module provides functions like Application, Document, and server_document 
to create and manage Bokeh applications.

Run the Bokeh server: Start the Bokeh server using the bokeh.server module. This module provides functions
like curdoc to access the current document, show to launch the server, and io to define and configure the 
server.

Update the plot: Use the Bokeh server's update mechanisms, such as callbacks or data streaming, to update
the plot in response to user interactions or data changes. This can involve modifying the plot's data 
sources, properties, or visual elements. """
```
